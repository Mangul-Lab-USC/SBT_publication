## Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as tkr
```

## Importing libraries

```python
color_dict=dict({'SBT':'blue','BWA':'orange'})
```

## Data reading

```python
data = pd.read_csv('../summary_data/cpu_memory_mapping.csv')
```

## Maximum memory usage

```python
sns.set_style("white")
sns.set_context("poster")

g=sns.catplot(x="method", 
              y="max_mem",
              data = data,
              palette=color_dict, 
              kind="bar",
              aspect=1.3)

import matplotlib.ticker as ticker

for ax in g.axes.flatten():
    ax.yaxis.set_major_formatter(tkr.FuncFormatter(lambda x, p: "{:.0f}".format(x/1024)))
    ax.yaxis.set_major_locator(ticker.MultipleLocator(1500))

g.set( ylabel='Maximum RAM (GB)')
g.set( xlabel='Pipeline')

sns.despine()

plt.savefig("../figures/memory.png", bbox_inches='tight')
plt.savefig("../figures/memory.pdf", bbox_inches='tight')
```

## CPU time

```python
sns.set_style("white")
sns.set_context("poster")

g=sns.catplot(x="method", 
              y="cpu_time",
              data = data,
              palette=color_dict, 
              kind="bar",
              aspect=1.3)

import matplotlib.ticker as ticker

for ax in g.axes.flatten():
    ax.yaxis.set_major_formatter(tkr.FuncFormatter(lambda x, p: "{:.0f}".format(x/60)))
    ax.yaxis.set_major_locator(ticker.MultipleLocator(1200))

g.set( ylabel='CPU time (min)')
g.set( xlabel='Pipeline')

sns.despine()

plt.savefig("../figures/time.png", bbox_inches='tight')
plt.savefig("../figures/time.pdf", bbox_inches='tight')
```
